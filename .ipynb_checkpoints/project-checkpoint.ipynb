```python
class NN(object):
    def __init__(self):
        pass

    def train(self, X, Y):
        hidden_layer_hight = 100
        output_layer_hight = 10

        w1 = np.random.randn(X.shape[1], hidden_layer_hight) * 0.01
        b1 = np.zeros((1, hidden_layer_hight))
        w2 = np.random.randn(hidden_layer_hight, output_layer_hight) * 0.01
        b2 = np.zeros((1, output_layer_hight))

        for i in range(30000):
            number_of_examples = 256
            data_batch = np.random.choice(len(X), number_of_examples, replace=False)
            x = X[data_batch]
            y = Y[data_batch]

            # forward pass
            hidden_layer = np.dot(x, w1) + b1    
            # relu
            hidden_layer = np.maximum(hidden_layer, 0, hidden_layer)
            # output
            scores = np.dot(hidden_layer, w2) + b2

            predictions = np.argmax(scores, axis=1)

            #result = y == predictions
            #print(len(result[result==True]) / number_of_examples)
            print('accuracy: %f' % ( np.mean(predictions == y) ))

            output_exp = np.exp(scores)
            output_exp_sum = output_exp.sum(axis=1)
            pk = np.divide(output_exp, output_exp_sum[:,np.newaxis])
            pk_y = pk[range(number_of_examples), y]

            loss = -np.log(pk_y)
            #print(np.mean(loss))

            # back propagation
            dldf = np.copy(scores)
            dldf[range(number_of_examples),y] -= 1

            dw2 = np.dot(hidden_layer.T, dldf)/number_of_examples
            db2 = np.sum(dldf, axis=0, keepdims=True)

            dhidden = np.dot(dldf, w2.T)
            dhidden[hidden_layer <= 0] = 0

            dw1 = np.dot(x.T, dhidden)/number_of_examples
            db1 = np.sum(dhidden, axis=0, keepdims=True)

            learning_rate = 0.0005
            w1 -= learning_rate * dw1
            b1 -= learning_rate * db1
            w2 -= learning_rate * dw2
            b2 -= learning_rate * db2
        
        self.w1 = w1
        self.b1 = b1
        self.w2 = w2
        self.b2 = b2
            
    def predict(self, X):
        # forward pass
        hidden_layer = np.dot(X, self.w1) + self.b1    
        # relu
        hidden_layer = np.maximum(hidden_layer, 0, hidden_layer)
        # output
        scores = np.dot(hidden_layer, self.w2) + self.b2

        predictions = np.argmax(scores, axis=1)
        
        return predictions
    
```

```python
#train(x[0:50000,:], y[0:50000])
nn = NN()
nn.train(x,y)
predictions = nn.predict(x[0:1000,:])
print('------------ test ------------')
print('accuracy: %f' % ( np.mean(predictions == y[0:1000]) ))
```
